```python
import os
import pandas as pd
import numpy as np

pathname = ""
#has 46 csv files with the top 50 songs with audio feature tags
years = ["2017","2018","2019"]
regions = ["us","gb", "ar", "at", "au", "be", "bg",
               "bo", "br", "ca", "ch", "cl", "co", "cy", "cz", "de",
               "dk", "es", "fi", "fr", "hk",
               "hn", "hu", "id", "ie", "is", "it", "jp", "lu", "lv",
               "mx", "my", "ni", "no", "pe",
               "ph", "pl", "pt", "py", "se", "sg", "sk", "sv", "tr", "tw", "uy"]
len(regions)
```

```python
filepath = "/Users/lalitha/Documents/worldhappinessmusic/countrycodes.csv"
df_codes = pd.read_csv(filepath)
countries_dict = {}
for i, j in df_codes.iterrows():
    country_code = df_codes["Country Code"][i].lower()
    if country_code in regions:
        countries_dict[country_code] = df_codes["Name"][i]

print(list(set(regions)-set(countries_dict.keys())))
countries_dict['jp'] = "Japan"
countries_dict['id'] = "Indonesia"
countries_dict['ca'] = "Canada"

countries_dict
```

```python
year = "2017"
happiness = pd.read_csv("/Users/lalitha/Documents/worldhappinessmusic/happiness_data/"+year+".csv")
happiness["Danceability"] = np.nan
happiness["Valence"] = np.nan
happiness["Energy"] = np.nan
happiness
```

```python

for year in years:
    print(year)
    
    happiness = pd.read_csv("/Users/lalitha/Documents/worldhappinessmusic/happiness_data/"+year+".csv")
    happiness["Danceability"] = np.nan
    happiness["Valence"] = np.nan
    happiness["Energy"] = np.nan
    happiness
    break
    for region in regions:
        print(region)
        country_name = countries_dict[region]
        filepath = "/Users/lalitha/Documents/worldhappinessmusic/top50_"+year+"_audiofeatures/top50stats_"+region+"_"+year+".csv"
        df = pd.read_csv(filepath)
        
        avg_danceability=df["Danceability"].mean()
        avg_valence=df["Valence"].mean()
        avg_energy = df["Energy"].mean()
        
        happiness["Danceability"] = avg_danceability
        happiness["Valence"] = avg_valence
        happiness["Energy"] = avg_energy
        

    happiness.dropna()
    break

happiness
                
    
    
```
