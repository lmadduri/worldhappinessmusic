```python
def add_statistics(region):
    print(region)
    #do once for each year
    ids = ",".join(ids_19[region])
    r = requests.get(BASE_URL + 'audio-features', headers=headers, params={'ids': ids})
    json_object = r.json()
    #a list of json objects
    audio_features = json_object['audio_features']
    # len(audio_features)
    danceability_stats = []
    valence_stats = []
    energy_stats = []
    tempo_stats = []
    for i in audio_features:
        danceability_stats.append(i["danceability"])
        valence_stats.append(i["valence"])
        energy_stats.append(i["energy"])

    filepath = "/Users/lalitha/Downloads/top50_2019/top50_"+region+"_2019.csv"
    df = pd.read_csv(filepath)
    df = df.assign(Danceability=danceability_stats, Valence=valence_stats, Energy=energy_stats)
    df.to_csv("top50stats_"+region+"_2019.csv",index=False)

for region in regions:
    add_statistics(region)
```
